```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#Tabloda eksik kalmış veriler var bunları dolduruyoruz
data_selected['Bisiklet_Tipi'].fillna(data_selected['Bisiklet_Tipi'].mode()[0], inplace=True)
data_selected['Fren_Tipi'].fillna(data_selected['Fren_Tipi'].mode()[0], inplace=True)
data_selected['Garanti_Suresi'].fillna(data_selected['Garanti_Suresi'].mode()[0], inplace=True)
data_selected['Jant'].fillna(data_selected['Jant'].mode()[0], inplace=True)
data_selected['Materyal'].fillna(data_selected['Materyal'].mode()[0], inplace=True)
data_selected['Renk'].fillna(data_selected['Renk'].mode()[0], inplace=True)
data_selected['Tekerlek_Sayisi'].fillna(data_selected['Tekerlek_Sayisi'].mode()[0], inplace=True)
data_selected['Vites'].fillna(data_selected['Vites'].mode()[0], inplace=True)
data_selected['Vites_Sistemi'].fillna(method='bfill', inplace=True)#çok az veri var kendisinden önce ve sonrakiyle doldurdum
data_selected['Vites_Sistemi'].fillna(method='ffill', inplace=True)
```

```python
#Tabloda eksik veri varlığını kontrol etme
print(data_selected['Bisiklet_Tipi'].isna().any())
print(data_selected['Fren_Tipi'].isna().any())
print(data_selected['Garanti_Suresi'].isna().any())
print(data_selected['Jant'].isna().any())
print(data_selected['Materyal'].isna().any())
print(data_selected['Renk'].isna().any())
print(data_selected['Tekerlek_Sayisi'].isna().any())
print(data_selected['Vites'].isna().any())
print(data_selected['Vites_Sistemi'].isna().any())
```

```python
#Verileri sayısallaştırmak için Label Encoder kullanma
from sklearn.preprocessing import LabelEncoder 
labelEncoder = LabelEncoder()
#fren tipi ile ilk deneme
print(data_selected["Fren_Tipi"].value_counts())
data_selected["Fren_Tipi"] = labelEncoder.fit_transform(data_selected["Fren_Tipi"])
print(data_selected["Fren_Tipi"].value_counts())
```

```python
#kalan kolonları sayısallaştırma
columns_to_encode = ["Bisiklet_Tipi", "Garanti_Suresi", "Jant", "Materyal", "Renk", "Vites", "Vites_Sistemi"]
labelEncoder = LabelEncoder()
for column in columns_to_encode:
    data_selected[column] = labelEncoder.fit_transform(data_selected[column])

data_selected.head()
```

```python
#Fiyat sütununu düzenleyip float a çevirme
def convert_price(price):
    price = price.replace(' TL', '')
    
    if '.' in price and ',' in price:
        price = price.replace(',', '')
    
    return float(price)


data_selected['Fiyat'] = data_selected['Fiyat'].apply(convert_price)

print(data_selected.head())
```

```python
#TRAIN VE TEST, ALGORİTMALAR KISMI
```

```python
# Gerekli Kütüphanelerin yüklenmesi
from sklearn.model_selection import learning_curve, RandomizedSearchCV, GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif, chi2
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
```

```python
# Algoritmamızda kullanmaya karar verdiğimiz verileri seçme
selected_columns = ["Fren_Tipi", "Jant", "Materyal", "Renk", "Vites"]
X = data_selected[selected_columns]
y= data_selected["Fiyat"]

# Eğitim ve test şeklinde ayırma %80 eğitim %20 test 
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

print('x_train :', x_train.shape)
print('x_test :', x_test.shape)
print('y_train :', y_train.shape)
print('y_test :', y_test.shape)
```

```python
#Bağımlı değişkenin bağımsız değişkenlerle olan doğrusal ilişki modelini oluşturan Linear Resgresyon modeli
from sklearn.linear_model import LinearRegression

# Linear Regression modelini oluşturma
lr = LinearRegression()
lr.fit(x_train, y_train)

#x bağımsız y bağımlı değişkenken tahmin etme kısmı 
y_predict_lr = lr.predict(x_test)

# mae, rmse, r^2 performans ölçekleriyle performansı değerlendir
#mae tahmin edilen ile gerçek arasındaki mutlak hata
mae_lr = mean_absolute_error(y_test, y_predict_lr)
#rmse hata değerlerinin karelerinin ortalamasının karekökü
rmse_lr = np.sqrt(mean_squared_error(y_test, y_predict_lr))
#r^2 ise hata kareler toplamının toplam varyansa bölünüp 1 den çıkarılmasıdır
r2_lr = r2_score(y_test, y_predict_lr)

print('Mean Absolute Error (Linear Regression):', mae_lr)
print('Root Mean Squared Error (Linear Regression):', rmse_lr)
print('Coefficient of Determination (R^2) (Linear Regression):', r2_lr)

# Grafik oluşturma kısmı
fig = plt.figure(figsize=(12, 6))
sonuclar = pd.DataFrame({'Tahmin Edilen': y_predict_dt, 'Gerçek': y_test})
sonuclar = sonuclar.reset_index()
sonuclar = sonuclar.drop(['index'], axis=1)
plt.plot(sonuclar)
plt.legend(['Gerçek', 'Tahmin Edilen'])
plt.title('Gerçek vs Tahmin Edilen')
plt.show()
```

```python
#Ağaç yapısıyla node ve leaf veriyi modelleyen algoritma Karar Ağacı
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(criterion='squared_error',splitter='best')
dt.fit(x_train,y_train)

y_predict_dt = dt.predict(x_test)

mae_dt = mean_absolute_error(y_test, y_predict_dt)
rmse_dt = np.sqrt(mean_squared_error(y_test, y_predict_dt))
r2_dt = r2_score(y_test, y_predict_dt)

print('Mean Absolute Error (Decision Tree):', mae_dt)
print('Root Mean Squared Error (Decision Tree):', rmse_dt)
print('Coefficient of Determination (R^2) (Decision Tree):', r2_dt)

fig = plt.figure(figsize=(12, 6))
sonuclar = pd.DataFrame({'Tahmin Edilen': y_predict_dt, 'Gerçek': y_test})
sonuclar = sonuclar.reset_index()
sonuclar = sonuclar.drop(['index'], axis=1)
plt.plot(sonuclar)
plt.legend(['Gerçek', 'Tahmin Edilen'])
plt.title('Gerçek vs Tahmin Edilen')
plt.show()
```

```python
# Regresyon katsayılarını küçültmeye ve sıfıra yaklaştırmaya yönelik bir düzenleme sağlayan algoritma Lasso 
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.001)
lasso.fit(x_train,y_train)
y_predict_lasso = lasso.predict(x_test) 

mae_lasso = mean_absolute_error(y_test, y_predict_lasso)
rmse_lasso = np.sqrt(mean_squared_error(y_test, y_predict_lasso))
r2_lasso = r2_score(y_test, y_predict_lasso)

print('Mean Absolute Error (Lasso):', mae_lasso)
print('Root Mean Squared Error (Lasso):', rmse_lasso)
print('Coefficient of Determination (R^2) (Lasso):', r2_lasso)

# Lasso İçin Tahmin Grafiği Çizelim
ffig = plt.figure(figsize=(12, 6))
sonuclar = pd.DataFrame({'Tahmin Edilen': y_predict_lasso, 'Gerçek': y_test})
sonuclar = sonuclar.reset_index()
sonuclar = sonuclar.drop(['index'], axis=1)
plt.plot(sonuclar)
plt.legend(['Gerçek', 'Tahmin Edilen'])
plt.title('Gerçek vs Tahmin Edilen')
plt.show()
```

```python
#Birden çok karar ağacını bir araya getirerek bir model oluşturan esemble(topluluk) öğrenme biçimi
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train,y_train)

y_predict_rf = rf.predict(x_test)

mae_rf = mean_absolute_error(y_test, y_predict_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_predict_rf))
r2_rf = r2_score(y_test, y_predict_rf)

print('Mean Absolute Error (Random Forest):', mae_rf)
print('Root Mean Squared Error (Random Forest):', rmse_rf)
print('Coefficient of Determination (R^2) (Random Forest):', r2_rf)

fig = plt.figure(figsize=(12, 6))
sonuclar = pd.DataFrame({'Tahmin Edilen': y_predict_rf, 'Gerçek': y_test})
sonuclar = sonuclar.reset_index()
sonuclar = sonuclar.drop(['index'], axis=1)
plt.plot(sonuclar)
plt.legend(['Gerçek', 'Tahmin Edilen'])
plt.title('Gerçek vs Tahmin Edilen')
plt.show()
```
